*  Here I have a bank data set in which I have to build a Decision Tree so that we can know who are all can repay the loan amount and who are all not.




```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import accuracy_score
from sklearn import preprocessing
```

```python
bank=pd.read_csv('/content/Decision_Tree_ Dataset(Bank Report).csv')
bank.head()
```

```python
bank.info()
```

```python
label_encoder=preprocessing.LabelEncoder()
bank['Result']=label_encoder.fit_transform(bank['Result'])
bank['Result'].unique() #1-is 'No' and 0 is 'yes'
```

```python
# now divide the data into dependent and indepentant variable
x=bank.iloc[:,1:5]
y=bank.iloc[:,0]
```

```python
#now create x_train,y_train and x_test,y_test
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=100)
```

```python
#Create a model
model=DecisionTreeClassifier(criterion='entropy',max_depth=3,random_state=100)
model.fit(X_train,y_train)
```

```python
#Function to predict
y_pred=model.predict(X_test)
print(y_pred)
```

```python
bank.head()
```

```python
tree.plot_tree(model)
```

```python
# checking the accuracy of the model
print('the accuracy of the model is',accuracy_score(y_test,y_pred)*100)
```

```python
bank.head()
```

```python
fn=['InitialPayment','LastPayment','CreditScore','HouseNumber']
cn=['Yes','No']
fig,axis=plt.subplots(nrows=1,ncols=1,figsize=(4,4),dpi=500)
tree.plot_tree(model,feature_names=fn,class_names=cn,filled=True)
```

```python
#predicting the value count for each category
pred=model.predict(X_test) # predicting the test on each data set
pd.Series(pred).value_counts() # counting the category
```

```python
# going to prepare a cross tab
pd.crosstab(y_test,pred) # getting the two way tabel to under stand the correct and wrong prediction
```

```python
#Acuarcy
np.mean(pred==y_test)
```
